S&P500 ('^GSPC')

```python
%matplotlib inline
import matplotlib.pyplot as plt
import matplotlib.finance as f
from matplotlib.dates import date2num
quotes = f.quotes_historical_yahoo_ochl('^GSPC', (2016, 2, 12), (2017, 2, 10), asobject=True, adjusted=True)
```

```python
quotes.date
```

```python
quotes.open
```

```python
quotes.close
```

```python
quotes.low
```

```python
data = [(date2num(q.date), q.open, q.close, q.high, q.low) for q in quotes]
data
```

```python
fig, ax = plt.subplots()
ax.grid()
ax.set_xlim(quotes[0][0], quotes[-1][0])
f.candlestick_ochl(ax, data, width=0.7, colorup='g', colordown='r', alpha=0.75)
fig.autofmt_xdate()
```
